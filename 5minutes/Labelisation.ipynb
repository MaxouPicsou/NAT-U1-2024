```python
import pandas as pd
import math
import socket
import matplotlib.pyplot as plt
```

```python
network_data = pd.read_csv("capture.csv")
action_data = pd.read_csv("action_file.csv")
action_data
```

```python
#===========================================================#
#==  Function to associate flows generated by each action ==#
#===========================================================#

# Create empty lists of actions and activities
action_list = [ math.nan ] * len(network_data)
activity_list = [ math.nan ] * len(network_data)

# Flow indices of an action
indices = []
# Flow indices list of all actions
flow_id_list = []
# Flow apearance order
flow_order = network_data.sort_values(by=["bidirectional_first_seen_ms"]).id.to_list()
# Action number
action_number = 1
# Last flow uncategorized
last_flow = 0

for timestamp_indice in range(len(action_data["Time(s)"]) - 1):
    
    # Time window between two actions 
    min_time = action_data.iloc[timestamp_indice]["Time(s)"]
    max_time = action_data.iloc[timestamp_indice + 1]["Time(s)"]
    
    # Print action details
    print("Action number", action_number, ":", action_data.iloc[action_number - 1]["Action"])
    print("between", min_time, "s and", max_time, "s")
    
    # Flow selection
    for flow_indice in flow_order[last_flow:]:
        
        flow_start_time = network_data.iloc[flow_indice]["bidirectional_first_seen_ms"]/1000

        # While flows are in window
        if min_time < flow_start_time < max_time:

            print("Flow number", flow_indice, "at",
                  network_data.iloc[flow_indice]["bidirectional_first_seen_ms"]/1000, "s",
                 "is in")#, action_data.iloc[timestamp_indice]["Activity"], "categorie")
            indices.append(flow_indice)

            # Update flow activity categorie
            action_list[flow_indice] = action_data.iloc[timestamp_indice]["Action"]

            # Check next flow
            last_flow += 1
            continue
        elif flow_start_time < min_time:
            last_flow += 1
            continue            
        else:
            break

    # If indices is empty, assigned NaN value
    if not indices:
        indices = math.nan
        
    # Adding flow indices list of the action in final flow list
    flow_id_list.append(indices)

    # Next action
    action_number += 1
    indices = []
    print()
    
```

```python
action_data = action_data.assign(Flows=pd.Series(flow_id_list))
action_data.to_csv("action_data_flows.csv")
action_data
```

```python
#=========================================================================#
#==  Function to associate activity granularity generated for each flow ==#
#=========================================================================#
def get_protocol_name(protocol_number_serie):
    result = [] 
    for i in protocol_number_serie:
        result.append(protocol_list[i])
        
    return result

def get_service_name(service_name_list):
    result = []
    for i in service_name_list:
        try:
            result.append(socket.getservbyport(i))
        except:
            result.append(math.nan)
            
    return result

    
PREFIXE = "IPPROTO_"
protocol_list = {}

# Generate protocol list number/name
for value, key in vars(socket).items():
    if PREFIXE in value:
        protocol_list[key] = value.replace(PREFIXE, "")

# Binary Granularity
network_data = network_data.assign(Binary=pd.Series(["Normal"] * len(network_data)))
# Protocol Granularity
network_data = network_data.assign(Protocol=pd.Series((get_protocol_name(network_data.protocol.to_list()))))
# Service Granularity
network_data = network_data.assign(Service=pd.Series((get_service_name(network_data.dst_port.to_list()))))
# Application Granularity
network_data = network_data.assign(Application=network_data.application_name)
# Webpage Granularity
network_data = network_data.assign(Webpage=network_data.requested_server_name)
# Action Granularity
network_data = network_data.assign(Action=pd.Series(action_list))
# Activity 
network_data = network_data.assign(Activity=network_data.application_category_name)

network_data.to_csv("network_data_flows.csv")

network_data
```
